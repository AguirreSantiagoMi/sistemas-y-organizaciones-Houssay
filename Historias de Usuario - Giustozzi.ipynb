# User Stories

### Como empresa de comidas:

* Como empresa de comidas, quiero poder gestionar eficientemente mi inventario a través de una aplicación para asegurarme de tener suficientes ingredientes para satisfacer la demanda de los clientes.

* Como empresa de comidas, deseo ofrecer a los clientes la posibilidad de realizar pedidos en línea a través de una aplicación móvil o un sitio web para aumentar la conveniencia y la accesibilidad.

* Como empresa de comidas, necesito una función de gestión de menús en la aplicación que me permita agregar, eliminar o modificar productos fácilmente para reflejar cambios en la oferta de alimentos.

### Como cliente de una empresa de comidas:

* Como cliente de una casa de comidas, quiero poder realizar pedidos en línea a través de una aplicación móvil para evitar largas esperas en el restaurante y tener una experiencia más conveniente y rápida.

* Como cliente de una casa de comidas, deseo poder personalizar mi pedido a través de la aplicación móvil o el sitio web para satisfacer mis preferencias dietéticas o gustos individuales, como agregar o eliminar ingredientes, elegir el nivel de picante, o solicitar extras.

* Como cliente de una casa de comidas, necesito tener acceso a promociones y ofertas especiales a través de la aplicación o el sitio web para disfrutar de descuentos exclusivos y ahorrar dinero en mis pedidos regulares.

### Como repartidor de una empresa de comidas:

* Como repartidor de comida, quiero recibir notificaciones en tiempo real sobre nuevos pedidos para poder prepararme para la entrega lo antes posible.

* Como repartidor de comida, necesito acceder a la información detallada de cada pedido, incluyendo la dirección de entrega, los artículos solicitados y las instrucciones especiales, para garantizar una entrega precisa y satisfactoria.

* Como repartidor de comida, deseo tener la capacidad de comunicarme fácilmente con el cliente a través de la aplicación en caso de preguntas o problemas relacionados con el pedido.

# Criterios de aceptación:

### Como empresa de comidas
* Gestión de inventario eficiente: La aplicación debe permitir a la empresa de comidas realizar un seguimiento preciso del inventario de ingredientes y productos en tiempo real. Debe ser capaz de mostrar cantidades disponibles, alertar sobre niveles bajos de existencias y permitir la actualización manual o automática del inventario después de cada venta.

* Procesamiento de pedidos en línea: La empresa de comidas debe ser capaz de recibir pedidos en línea a través de la aplicación y/o sitio web. Los criterios de aceptación incluyen la recepción instantánea de pedidos, la capacidad de visualizar y confirmar cada pedido, y la integración con sistemas de pago seguro para procesar transacciones de manera eficiente.

* Seguimiento de entregas: La aplicación debe proporcionar a la empresa de comidas una herramienta para rastrear el estado de las entregas en tiempo real. Debe permitir asignar repartidores a pedidos específicos, mostrar la ubicación actual de los repartidores y proporcionar actualizaciones automáticas sobre el estado de las entregas, como "en camino" o "entregado".

### Como cliente de una empresa de comidas

* Interfaz de usuario intuitiva: La aplicación o el sitio web de la casa de comidas debe ser fácil de navegar y utilizar, con un diseño intuitivo que permita a los clientes encontrar rápidamente los productos, personalizar sus pedidos y completar el proceso de compra sin confusiones.

* Procesamiento de pedidos eficiente: Los pedidos realizados a través de la aplicación o el sitio web deben ser procesados de manera rápida y precisa. Los criterios de aceptación incluyen una confirmación inmediata del pedido, actualizaciones en tiempo real sobre el estado del pedido (como "en preparación" o "en camino"), y notificaciones de entrega una vez que el pedido ha sido completado.

* Acceso a promociones y descuentos: La aplicación o el sitio web de la casa de comidas deben proporcionar a los clientes acceso fácil a promociones, descuentos y ofertas especiales. Los criterios de aceptación incluyen la visualización clara de las promociones disponibles, la aplicación automática de descuentos durante el proceso de compra y la capacidad de canjear cupones promocionales.

### Como repartidor de una empresa de comidas

* Notificaciones en tiempo real: El repartidor debe recibir notificaciones instantáneas en su dispositivo móvil cada vez que se genere un nuevo pedido en la plataforma.

* Acceso a detalles del pedido: El repartidor debe poder acceder a todos los detalles relevantes del pedido, incluyendo la dirección de entrega, los artículos solicitados, las instrucciones especiales y el nombre del cliente.

* Función de comunicación integrada: Debe haber una función de mensajería dentro de la aplicación que permita al repartidor comunicarse directamente con el cliente en caso de preguntas o problemas relacionados con el pedido.
